```python
#Import pandas
import pandas as pd
```

```python
#Reads in CSV File and removes unnecessary rows
df = pd.read_csv("GlobalLandTemperaturesByCountry.csv", index_col = ['dt'], usecols=['dt', 'AverageTemperature', 'Country'])

df
```

```python
#Date Time Conversion
df.index = pd.to_datetime(df.index)
```

```python
#Removes NaN values
df = df.dropna(how='any')
df
```

```python
df = df.loc[df['Country'].isin(['Honduras', 'Haiti','Zimbabwe','Philippines','China', 'United States', 'Japan', 'United Kingdom'])]
df
```

```python
# filtered_df = df.loc[(df['dt'] >= '1960-01-01') & (df['dt'] < '2020-09-15')]
# df = pd.DatetimeIndex("1960-01-01","2020-01-01")
df = df.loc['1960':'2013']
df
```

```python
df = df.loc[pd.DatetimeIndex(df.index).month == 8]
df
```

```python
#Export
df = df.to_csv('completedTemperature.csv', index = True)
```
